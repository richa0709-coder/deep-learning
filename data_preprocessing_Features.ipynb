```
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```
##tokenization - paragraph - sentences - words
paragraph
nltk.download('punkt')
nltk.download('punkt_tab')
sentences=nltk.sent_tokenize(paragraph)
```

```
sentences
```

```
type(sentences)
len(sentences)
```

```
stemmer=PorterStemmer()
```

```
stemmer.stem('drinking')
```

```
from nltk.stem import WordNetLemmatizer
```

```
lemmatizer=WordNetLemmatizer()
```

```
nltk.download('wordnet')
lemmatizer.lemmatize('history')
```

```
import re
```

```
corpus=[]
```

```
for i in range(len(sentences)):
  review = re.sub(r'\[.*?\]', '', sentences[i]) # remove citations like [a], [b], etc.
  review = re.sub(r'[^a-zA-Z\s]', '', review) # remove special characters except letters and spaces
  review = review.lower()
  #review.split()
  corpus.append(review)
```

```
print(corpus)
```

```
import nltk
nltk.download('stopwords')
corpus
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```
##Apply stopwords, lemmatize
import re
corpus=[]
for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  review=review.split()
  review=[lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review=' '.join(review)
  corpus.append(review)
```

```
corpus
```

```
## Bag of words

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(corpus)
```

```
cv.vocabulary_
```

```
X[0].toarray()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer()
X=cv.fit_transform(corpus)
```
